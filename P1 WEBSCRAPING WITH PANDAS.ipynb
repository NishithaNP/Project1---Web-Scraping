```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# creating list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames = []
# looping through leagues, read multiple csv and append it into a list
for i in leagues:
    df = pd.read_csv(root + '2122/' + i + '.csv',encoding='unicode_escape')
    frames.append(df)
```

```python
#MULTIPLE SEASONS
# creating list of leagues
#https://www.football-data.co.uk/mmz4281/2021/E0.csv
#https://www.football-data.co.uk/mmz4281/1920/E0.csv 
leagues = ['E0','E2','E3']
frames = []
# looping through leagues, read multiple csv and append it into a list
for i in leagues:
    for j in range(15,21):
        df = pd.read_csv(root + str(j)+str(j+1) +'/'+ i + '.csv')
        df.insert(1,'season',j)
        frames.append(df)
```

# ORGANIZE ALL THE DATA IN A DICTIONARY

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['German Bundesliga']
```

```python
import pandas as pd
dict_historical_data={}

leagues = ['E0','E2','E3']
# looping through leagues, read multiple csv and append it into a list
for i in dict_countries:
    frames = []
    for j in range(15,21):
        df = pd.read_csv(root + str(j)+str(j+1) +'/'+ dict_countries[i] + '.csv')
        df.insert(1,'season',j)
        frames.append(df)
    df_concat=pd.concat(frames)
    dict_historical_data[i]=df_concat
```
